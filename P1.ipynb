**Questão 3:**
A fórmula de Leibniz para o número $\pi$ é dada pela série infinita $\frac{\pi}{4} = \sum_{n=0}^{\infty}\frac{(-1)^n}{2n+1}$. Apresente um programa em Python para obter uma aproximações para $\pi$ usando 50 termos da série. Calcule os erros absoluto e relativo.

```python
s = 0
s_ex = np.pi / 4
for n in range(50):
    s += (-1)**n/(2*n+1)
    e = np.abs(s - s_ex)
    e_r = np.abs(s - s_ex) / np.abs(s_ex)
    print(s, e, e_r)
```

```python
print(s*4)
```

```python
print('Valor exato: ', s_ex)
print('Valor aproximado: ', s)
print('Erro absoluto: ', e)
print('Erro relativo (%): ', e_r*100)
```

**Questão 4.** Para cada uma das equações a seguir faça um gráfico para localizar a(s) raiz(es) e utilize o método da bisseção e o método do ponto fixo (se possível) para determinar todas as soluções das equações abaixo com precisão de $10^{-5}$:

**a)** $x^2-4 x+4-\ln x=0 \quad$ 

**b)** $x+1-2 \operatorname{sen} (\pi x)=0 \quad$  

Comente o que você observa em relação à convergência de cada método.

```python
f0 = lambda x: x**2-4*x+4-np.log(x)
x = np.linspace(1e-6, 5, 500)
plt.plot(x, f0(x))
plt.grid()
plt.show()
```

```python
err = 10
a = 1
b = 2
x_ant = 2
f = f0
i = 1
while err > 1e-5:
    x = (a+b)/2.0
    if f(x)*f(a) < 0:
        b = x
    else:
        a = x
    err = abs(x - x_ant)/abs(x)
    x_ant = x
    print(i)
    print ("x = %.4f"%x,"| Erro = %.4f"% err)
    i+=1
```

```python
phi = lambda x: x**2/4 + 1 - np.log(x)/4
x = np.linspace(1e-6, 5, 500)
plt.plot(x, phi(x))
plt.grid()
plt.show()
```

```python
x0 = 1.5

# Inicializando as variáveis
err = 10.0
x_ant = x0
i = 0

while err>1e-5:
    x = phi(x_ant)
    err = abs(x-x_ant)/abs(x)
    x_ant = x
    i=i+1
    print (i, "| x=%.9f"%x, "| err=%.9f"%err)
```

```python
f = lambda x: np.log(x) - np.sin(x)
x = np.linspace(1e-5, 10, 500)
plt.plot(x, f(x))
plt.grid()
plt.show()
```

```python
df = lambda x: 1./x - np.cos(x)  
x = 2
x_ant = x
err = 10.0
i=1
while err>1e-5:
    x = x - f(x)/df(x)
    err = abs(x-x_ant)/abs(x)
    x_ant = x
    print (i," | x=%.6f"%x, "| err=%.6f"%err)
    i+=1
```

```python
err=10.
x0=2.2
x1=2

while err> 0.0001:
    x = (x0*f(x1)-x1*f(x0))/(f(x1)-f(x0))
    err = abs(x-x1)/abs(x)   # calcula o erro relativo
    x0 = x1
    x1 = x
    print ("x=%.5f"%x, ", err=%.5f"%err)
```
