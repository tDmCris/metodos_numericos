A fórmula de bhaskara é $ x = \frac{-b \pm \sqrt{b^2-4ac}}{2a}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constants
c = 12.5
m = 68.1
g = 9.81

# Define the initial conditions
v0 = 0
t0 = 0

# Define the time step
dt = 2
# Define the number of time steps
N = 6
# Create arrays to store the values of t and v
t = np.zeros(N+1)
v = np.zeros(N+1)
# Set the initial values
t[0] = t0
v[0] = v0
# Use the forward Euler method to solve
for i in range(N):
    t[i+1] = t[i] + dt
    v[i+1] = v[i] + dt * (g - c/m * v[i])
# Print the results
for i in range(N+1):
    print(f"t = {t[i]:.2f} s, v = {v[i]:.2f} m/s")
# Plot the results
plt.plot(t, v)
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.grid()
plt.show()
```
